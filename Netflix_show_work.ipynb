```python
import dtale
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
import numpy as np
```

Read CSV from Kaggle

```python
show_file = "Resources/netflix_titles.csv"
show_df = pd.read_csv(show_file)
show_df.head()
```

```python
new_show_df=show_df[["show_id", "type","title","release_year","rating"]]
new_show_df
```

Read csv from data.world

```python
user_rating_file = "Resources/netflix_stuff.csv"
user_rating_df = pd.read_csv(user_rating_file)
user_rating_df.head()
```

```python
new_user_rating_df=user_rating_df[["title","release year","rating","user rating score","user rating size"]]
new_user_rating_df
```

```python
clean_user_rating_df=new_user_rating_df.rename(columns={"release year": "release_year",
                                                          "user rating score": "user_rating_score",
                                                          "user rating size": "user_rating_size"})
clean_user_rating_df.head()
```

```python
user_rating_copy_df=clean_user_rating_df.copy()
user_rating_copy_df
```

```python
null_rating_df=user_rating_copy_df[user_rating_copy_df['user_rating_score'].isnull()]
null_rating_df.head()
```

```python
not_null_rating_df=user_rating_copy_df[user_rating_copy_df['user_rating_score'].notnull()]
not_null_rating_df
```

```python
# not_null_groupby_rating_df=not_null_rating_df.groupby('rating')
# not_null_groupby_rating_df.first()
```

Merged the cleaned rating data with the dataset from Kaggle

```python
merge_df = pd.merge(not_null_rating_df, new_show_df, on="title", how="outer")
merge_df
```

```python
merge_df.drop_duplicates()
```

Cleaned the data to get rid of any null values in user rating column.

```python
with_rating_score_merge_df=merge_df[merge_df['user_rating_score'].notnull()]
with_rating_score_merge_df
```

```python
dtale.show(with_rating_score_merge_df, ignore_duplicate=True)
```

```python
cleaned_movie_list_df= with_rating_score_merge_df.loc[(with_rating_score_merge_df['rating_x']=='G')|
                                                      (with_rating_score_merge_df['rating_x']=='NR')|
                                                      (with_rating_score_merge_df['rating_x']=='PG')|
                                                      (with_rating_score_merge_df['rating_x']=='PG-13')|
                                                      (with_rating_score_merge_df['rating_x']=='R')
                                                     ]
cleaned_movie_list_df=cleaned_movie_list_df.reset_index(drop=True)
dtale.show(cleaned_movie_list_df, ignore_duplicate=True)
```
