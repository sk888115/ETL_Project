```python
import dtale
import pandas as pd
from sqlalchemy import create_engine
import matplotlib.pyplot as plt
```

```python
#Call dataset 1 from Kaggle
show_file = "Resources/netflix_titles.csv"
show_df = pd.read_csv(show_file)
show_df.head()
```

```python
#Took the necessary colunms for dataset 1
new_show_df=show_df[["type","title","release_year","rating"]]
new_show_df
```

```python
#Call dataset 2 from data world
user_rating_file = "Resources/netflix_stuff.csv"
user_rating_df = pd.read_csv(user_rating_file)
user_rating_df.head()
```

```python
#Took the necessary colunms for dataset 2
new_user_rating_df=user_rating_df[["title","release year","rating","user rating score","user rating size"]]
new_user_rating_df
```

```python
# Renamed some columns in dataset 2, prepared for merge
clean_user_rating_df=new_user_rating_df.rename(columns={"release year": "release_year",
                                                          "user rating score": "user_rating_score",
                                                          "user rating size": "user_rating_size"})
clean_user_rating_df.head()
```

```python
# Pulled out the data that has null value in user_rating_column 
not_null_rating_df=clean_user_rating_df[clean_user_rating_df['user_rating_score'].notnull()]
not_null_rating_df
```

```python
# Merged data set 1 & cleaned data set 2
merge_df = pd.merge(not_null_rating_df, new_show_df, on="title", how="outer")
merge_df
```

```python
# Drop duplicates
merge_df.drop_duplicates()
```

```python
# Only pulled out the data with User Rating Score
with_rating_score_merge_df=merge_df[merge_df['user_rating_score'].notnull()]
with_rating_score_merge_df
```

```python
with_rating_score_merge_groupby_df=with_rating_score_merge_df.groupby(['rating_x'])
with_rating_score_merge_groupby_df.count()
```

```python
with_rating_score_merge_groupby_df=with_rating_score_merge_df.groupby(['rating_y'])
with_rating_score_merge_groupby_df.count()
```

```python
#Group decided to use "relase_year_x" & "rating_x"
#"rating_x" had more variety
#"release_year_x" is with "rating_x"
cleaned_merge_df=with_rating_score_merge_df[["title","release_year_x","rating_x","user_rating_score","user_rating_size", "type"]]

cleaned_merge_df
```

```python
cleaned_merge_df= cleaned_merge_df.rename(columns={"release_year_x": "release_year",
                                  "rating_x": "rating"})

cleaned_merge_df
```

```python
# Extract the list that has movie rating (G, NR, PG, PG-13, R) 
cleaned_movie_list_df= cleaned_merge_df.loc[(cleaned_merge_df['rating']=='G')|
                                            (cleaned_merge_df['rating']=='NR')|
                                            (cleaned_merge_df['rating']=='PG')|
                                            (cleaned_merge_df['rating']=='PG-13')|
                                            (cleaned_merge_df['rating']=='R')
                                                     ]
cleaned_movie_list_df=cleaned_movie_list_df.reset_index(drop=True)
cleaned_movie_list_df
#dtale.show(cleaned_movie_list_df, ignore_duplicate=True)
```

```python
#Replace null values in Type with with 'Movie' 
cleaned_movie_list_df["type"].fillna("Movie", inplace = True)
cleaned_movie_list_df
cleaned_movie_list_df=cleaned_movie_list_df.reset_index(drop=True)
cleaned_movie_list_df
#dtale.show(cleaned_movie_list_df, ignore_duplicate=True)
```

```python
# # Reference to figure out TV Shows in the movie list
# tvshow_in_movie_list_df=cleaned_movie_list_df.loc[(cleaned_movie_list_df['type']=='TV Show')  ]
# tvshow_in_movie_list_df

# Reference to figure out how many TV Shows were in the movie list
tvshow_in_movie_list_df=cleaned_movie_list_df.loc[(cleaned_movie_list_df['type']=='TV Show')  ]
tvshow_in_movie_list_df=tvshow_in_movie_list_df.reset_index(drop=True)
tvshow_in_movie_list_df
#dtale.show(tvshow_in_movie_list_df, ignore_duplicate=True)
```

```python
#Changed the type to Movie as "The Jungle Book" & "Goosebumps" & "Dope" are movie.
cleaned_movie_list_df.loc[(cleaned_movie_list_df['title']=='The Jungle Book')|
                                               (cleaned_movie_list_df['title']=='Goosebumps')|
                                               (cleaned_movie_list_df['title']=='Dope')]
cleaned_movie_list_df['type']=cleaned_movie_list_df['type'].replace(['TV Show'], 'Movie')


#Changed the rating to TV show rating as "Hunter x Hunter" & "The 100" are TV shows.
cleaned_movie_list_df.loc[(cleaned_movie_list_df['title']=='The 100')|
                                               (cleaned_movie_list_df['title']=='Hunter X Hunter (2011)')]
cleaned_movie_list_df['rating']=cleaned_movie_list_df['rating'].replace(['NR'], 'TV-14')



cleaned_movie_list_df=cleaned_movie_list_df.reset_index(drop=True)
cleaned_movie_list_df
#dtale.show(cleaned_movie_list_df, ignore_duplicate=True)
```

```python
#TV Show List
cleaned_tvshow_list_df= cleaned_merge_df.loc[(cleaned_merge_df['rating']=='TV-14')|
                                              (cleaned_merge_df['rating']=='TV-G')|
                                              (cleaned_merge_df['rating']=='TV-MA')|
                                              (cleaned_merge_df['rating']=='TV-PG')|
                                              (cleaned_merge_df['rating']=='TV-Y')|
                                              (cleaned_merge_df['rating']=='TV-Y7')|
                                              (cleaned_merge_df['rating']=='TV-Y7-FV')                                                       
                                                     ]
cleaned_tvshow_list_df=cleaned_tvshow_list_df.reset_index(drop=True)
#dtale.show(cleaned_tvshow_list_df, ignore_duplicate=True)
```

```python
#Replace null values in Type with with 'Movie' 
cleaned_tvshow_list_df["type"].fillna("TV Show", inplace = True)
cleaned_tvshow_list_df=cleaned_tvshow_list_df.reset_index(drop=True)
#dtale.show(cleaned_tvshow_list_df, ignore_duplicate=True)
```

```python
# Reference to figure out how many movies were in the TV Show list
movie_in_tv_list_df=cleaned_tvshow_list_df.loc[(cleaned_tvshow_list_df['type']=='Movie')  ]
movie_in_tv_list_df=movie_in_tv_list_df.reset_index(drop=True)
#dtale.show(movie_in_tv_list_df, ignore_duplicate=True)

movie_in_tv_list_df.drop_duplicates()
```

```python
cleaned_tvshow_list_df.drop(cleaned_tvshow_list_df.loc[cleaned_tvshow_list_df['type']=='Movie'].index, inplace=True)
cleaned_tvshow_list_df

cleaned_tvshow_list_df=cleaned_tvshow_list_df.reset_index(drop=True)
dtale.show(cleaned_tvshow_list_df, ignore_duplicate=True)
```

```python
# Exporting cleaned dataset
```

```python
# Dropped duplicates
final_tv_show=cleaned_tvshow_list_df.drop_duplicates()
final_tv_show
```

```python
# Dropped duplicates
final_movie_df=cleaned_movie_list_df.drop_duplicates()
final_movie_df
```

```python
#connect to local database
connection_string="postgres:postgres@localhost:5432/netflix_db"

engine = create_engine(f'postgresql://{connection_string}')
```

```python
#check for tables
engine.table_names()
```

```python
#use pandas to load csv converted df to database
final_tv_show.to_sql(name='tv_show', con=engine, if_exists='append', index=False)
```

```python
#use pandas to load csv converted df to database
final_movie_df.to_sql(name='movie', con=engine, if_exists='append', index=False)
```
